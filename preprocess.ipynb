```python
import pandas as pd

def preprocess(file):
    import codecs
    txt=''
    with codecs.open(f"data_txt/{file}.txt", encoding='utf-8') as f:
        for line in f:
            #print(repr(line))
            txt+=line
    list_question = txt.replace('\r','').replace('\t',' ').replace(' (Une seule réponse est exacte)','')
    list_question = list_question.replace(' (Deux réponses sont exactes)','')
    list_question = list_question.replace(' (Trois réponses sont exactes)','')
    list_question = list_question.split('\n')
    list_question = list(filter(lambda a: a != '', list_question))
    list_question = list(filter(lambda a: a != ' ', list_question))


    dict_qcm = {}
    num_qcm = 0
    for items in list_question:
        if 'QCM' in items[0:4]:
            num_qcm = num_qcm + 1
            dict_qcm[num_qcm] = {
                'Question':items[8:],
                'A':'',
                'B':'',
                'C':'',
                'D':'',
                'E':'',
                'Reponse':''
            }
            continue

        if 'A.' in items:
            dict_qcm[num_qcm]['A']=items
        elif 'B.' in items:
            dict_qcm[num_qcm]['B']=items
        elif 'C.' in items:
            dict_qcm[num_qcm]['C']=items
        elif 'D.' in items:
            dict_qcm[num_qcm]['D']=items
        elif 'E.' in items:
            dict_qcm[num_qcm]['E']=items
        elif 'Réponse :' in items:
            dict_qcm[num_qcm]['Reponse']=items
        else:
            print('---------------')
            print('START ERROR')
            print('Error',dict_qcm)
            print('---------------')
            print(items)
            print('END ERROR')
            print('---------------')

            pass

    import pandas as pd
    df = pd.DataFrame(dict_qcm).T
    df['Reponse_preprocess'] = df['Reponse'].str.replace('Réponse : ','')
    def get_reponse(row):
        reponse=''
        if len(row['Reponse_preprocess'].split(' '))==1:
            if row['Reponse_preprocess'].isupper() or c!= ' ':
                return row['Reponse_preprocess']
            else: 
                return None

        for i,c in enumerate(row['Reponse_preprocess']):
            if i==0 and not c.isupper() :
                raise ValueError
            if c.isupper() or c!= ' ':
                reponse=reponse + c
            else: 
                return reponse

    def split(row): 
        return [char for char in row['Reponse_preprocess']]  

    df['Reponse_preprocess'] = df.apply(get_reponse, axis=1)
    df['Reponse_preprocess'] = df.apply(split, axis=1)
    df['file'] = file
    return df
```

```python
df_list=[]
for file in ['chapitre1','chapitre2','chapitre3','chapitre4','chapitre5','institutions','economie']:
    print('Preprocess file :',file)
    df = preprocess(file)
    df_list.append(df)
    df.to_csv(f'data_csv/{file}.csv')
    print('DONE ',file)
df_concat = pd.concat(df_list)
df_concat.to_csv('data_csv/all_qcm.csv', index=False)
df = pd.read_csv('data_csv/all_qcm.csv')
df.head()
```
